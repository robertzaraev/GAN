```python
import cv2
import os
import glob
import numpy as np
import random 
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
image_paths = []
for image in os.listdir('avatars/'):
    image_paths.append('avatars/' + image)
image_paths
```

```python
img = cv2.imread(image_paths[0], 1)
img.shape
```

```python
d_first = {}
d_second = {}
d_third = {}
img = cv2.imread(image_paths[0], 1)
for i, val_x in enumerate(img[:,:,0].reshape(-1)):
        d_first[i] = {}
        d_second[i] = {}
        d_third[i] = {}
```

```python
for avatar in tqdm(image_paths):
    img = cv2.imread(avatar, 1)
    for dim in range(0,3):
        for i, val_x in enumerate(img[:,:,dim].reshape(-1)):
            if dim == 0:
                if val_x not in d_first[i].keys():
                    d_first[i][val_x] = 1
                else:
                    d_first[i][val_x] += 1
            elif dim == 1:
                if val_x not in d_second[i].keys():
                    d_second[i][val_x] = 1
                else:
                    d_second[i][val_x] += 1
            elif dim == 2:
                if val_x not in d_third[i].keys():
                    d_third[i][val_x] = 1
                else:
                    d_third[i][val_x] += 1
```

```python
d_third
```

```python
d_first_proba = {}
d_second_proba = {}
d_third_proba = {}
for key in d_first.keys():
    d_first_proba[key] = []
    d_second_proba[key] = []
    d_third_proba[key] = []
    for ind, val in enumerate(d_first[key].keys()):
        d_first_proba[key].append(d_first[key][val]/ sum(d_first[key].values()))
    for ind, val in enumerate(d_second[key].keys()):
        d_second_proba[key].append(d_second[key][val]/ sum(d_second[key].values())) 
    for ind, val in enumerate(d_third[key].keys()):
        d_third_proba[key].append(d_third[key][val]/ sum(d_third[key].values())) 
```

```python
first_result = []
second_result = []
third_result = []
for key in d_first.keys():
    first_result.extend(random.choices(list(d_first[key].keys()),d_first_proba[key]))
    second_result.extend(random.choices(list(d_second[key].keys()),d_second_proba[key]))
    third_result.extend(random.choices(list(d_third[key].keys()),d_third_proba[key]))
```

```python
first_result = np.array(first_result).reshape(560, 528,1)
second_result = np.array(second_result).reshape(560, 528,1)
third_result = np.array(third_result).reshape(560, 528,1)
```

```python
final = np.concatenate([first_result, second_result, third_result], -1)
```

```python
cv2.imshow('image', final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
filename = 'image.jpg'
```

```python
cv2.imwrite(filename, final)
```
